```python
# Importing statements 
import pandas # To handle dataset
from sklearn.model_selection import train_test_split # To split the dataset into training and testing sets
from sklearn.linear_model import LinearRegression # To create a linear regression model
import pickle # To save the model to disk

salary_dataset = pandas.read_csv("Salary_Data.csv") # Reading the dataset from a CSV file and storing it in a variable called salary_dataset

# Extracting the "YearsExperience" and "Salary" columns from the dataset and storing them in separate variables called years_of_experiance and salaries, respectively. The double square brackets are used to ensure that the result is a DataFrame rather than a Series.
years_of_experiance = salary_dataset[["YearsExperience"]]
salaries = salary_dataset[["Salary"]]

# Splitting the dataset into training and testing sets using the train_test_split function from scikit-learn. The test_size parameter is set to 0.2, which means that 20% of the data will be used for testing, and the random_state parameter is set to 0 to ensure reproducibility of the results.
x_train, x_test, y_train, y_test = train_test_split(years_of_experiance, salaries, test_size=0.2, random_state=0)

regressor = LinearRegression()
regressor.fit(x_train, y_train)

w = regressor.coef_ #[[9312.57512673]]  weigtage
b = regressor.intercept_ # [26780.09915063] bias value

r2 = regressor.score(x_test, y_test) # 0.988169515729126 Better model, 98%

file_name = "salary-pred-model.sav"
pickle.dump(regressor, open(file_name, "wb")) # Saving the model to disk

ai_model = pickle.load(open(file_name, "rb")) # Loading the model from disk

salary_predicted = ai_model.predict([[15]]) # [[166468.72605157]]

```
