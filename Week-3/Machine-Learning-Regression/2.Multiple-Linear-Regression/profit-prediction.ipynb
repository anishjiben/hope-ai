```python
# Importing statements 
import pandas # To handle dataset
from sklearn.model_selection import train_test_split # To split the dataset into training and testing sets
from sklearn.linear_model import LinearRegression # To create a linear regression model
import pickle # To save the model to disk

# Load the dataset
dataset = pandas.read_csv('50_Startups.csv')

# Convert categorical variable 'State' into numerical using one-hot-encoding
dataset = pandas.get_dummies(dataset,columns=['State'], drop_first=True) 

# Split Independent and Dependent variables
dependent_variable = dataset[['Profit']] # Target variable
independent_variables = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(independent_variables, dependent_variable, test_size=0.2, random_state=0)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Evaluate the model
score = model.score(X_test, y_test) #  0.9347068473282423

# Save the model to disk
file_name = 'profit-prediction-model.sav'
pickle.dump(model, open(file_name, 'wb'))


ai_model = pickle.load(open(file_name, 'rb')) # Load the model from disk
profit = ai_model.predict([[165349, 136897, 471784, 0, 1]]) # Make predictions using the loaded model

print(profit)
```
