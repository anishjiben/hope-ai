```python
# 4
number = input("Enter your password: ")

def is_prime(num):
    if num<2:
        return False
    else:
        for i in range(2,num):
            if num%i==0:
                return False
    return True

print(is_prime(int(number)))
```

```python
# 5
def factorial(num):
    if num == 1:
        return 1
    else:
        return num * factorial(num-1)
print(factorial(10))
```

```python
# Scenario 6
x=407

def is_armstrong(num):
    no_of_digits = len(str(num))
    num_list = list(map(lambda x: int(x)**no_of_digits, str(num)))
    return sum(num_list) == num

print(is_armstrong(x))
```

```python
# Scenario 9
sentence = " Hi, i am anish jiben"
words = sentence.split()
longest_word = words[0]

for word in words:
    if len(word) > len(longest_word):
        longest_word = word

print(longest_word)
```

```python
# Scenario 10

def is_anagram(word_1, word_2):
    str_1 = list(word_1)
    str_2 = list(word_2)
    str_1.sort()
    str_2.sort()
    if str_1== str_2:
        return True
    return False 


print(is_anagram("anish", "shina"))
```
