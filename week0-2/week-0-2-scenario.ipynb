```python
#Scenario 1
user_age=int(input("Please enter your age: "))
if user_age < 18:
    print("You are a minor.")
else:    
    print("Eligible to vote")
```

```python
# Largest number in a list
numbers = [3, 7, 2, 9, 5]
largest_number = max(numbers)
print("The largest number in the list is:", largest_number)
```

```python
# 10% bonus on salary
salary = float(input("Please enter your current salary: "))
if salary > 50000:
    bonus = salary * 0.10
    new_salary = salary + bonus
    print("Your new salary after a 10% bonus is:", new_salary)
else:
    print("No bonus applied. Your salary remains:", salary)
```

```python
# Odd or Even
number = int(input("Please enter a number: "))
if number % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")
```

```python
# Reverse a string
input_string = input("Please enter a string: ")
reversed_string = input_string[::-1]
print("The reversed string is:", reversed_string)
```

```python
# 20% discount on total bill
total_bill = float(input("Please enter the total bill amount: "))
if total_bill > 100:
    discount = total_bill * 0.20
    discounted_bill = total_bill - discount
    print("Your total bill after a 20% discount is:", discounted_bill)
else:
    print("No discount applied. Your total bill remains:", total_bill)
```

```python
# Leap year check
year = int(input("Please enter a year: "))

if year % 400 == 0:
    print(f"{year} is a leap year.")
elif year % 100 == 0:
    print(f"{year} is not a leap year.")
elif year % 4 == 0:
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
# Filter even numbers
numbers = [4,3,5,6,7,8,9]
even_numbers = [num for num in numbers if num % 2 == 0]
print("Even numbers:", even_numbers)
```
