```python
from pathlib import Path
import pandas as pd
from sklearn import tree
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split


data_path = Path("../datasets/insurance_pre.csv")
dataset = pd.read_csv(data_path)

sex_map = {'male': 0, 'female': 1}
smoker_map = {'yes': 1, 'no': 0}

# Pre process the dataset by mapping categorical variables to numerical values
dataset['sex'] = dataset['sex'].map(sex_map)
dataset['smoker'] = dataset['smoker'].map(smoker_map)

# Define input features and target variable
input_features = dataset[['age', 'sex', 'bmi', 'children', 'smoker']]
target = dataset[['charges']]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(input_features, target, test_size=0.2, random_state=0)


# Support Vector Regression
svr_model = SVR(kernel="sigmoid")
svr_model.fit(X_train, y_train)
y_pred = svr_model.predict(X_test)
svr_score = r2_score(y_test, y_pred)


print(f"Support Vector Regression R^2 Score: {svr_score}")






```
